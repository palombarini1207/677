```python
#2 - Contagem regressiva, o usuario digita um numero e o programa do zero os numeros pares
# Usuário digita um número
numero = int(input("Digite um número: "))

# Mostra os números pares do zero até o número digitado
for i in range(0, numero + 1, 2):
    print(i)
```

```python
#3 - Numeros pares, o usuario digita e o programa imrpimi os numeros pares
# Usuário digita um número
numero = int(input("Digite um número: "))

# Imprime os números pares de 0 até o número digitado
for i in range(0, numero + 1):
    if i % 2 == 0:
        print(i)
```

```python
#4 - Numeros impares, o usuario digita um numero e o programa imprimi os numeros impares
# Usuário digita um número
numero = int(input("Digite um número: "))

# Imprime os números ímpares de 0 até o número digitado
for i in range(0, numero + 1):
    if i % 2 != 0:
        print(i)
```

```python
#5 - Soma, o usuario digita um numero e o programa soma os numeros dp zero ate o numero
# Usuário digita um número
numero = int(input("Digite um número: "))

# Variável para armazenar a soma
soma = 0

# Soma os números de 0 até o número digitado
for i in range(0, numero + 1):
    soma += i

# Mostra o resultado
print("A soma é:", soma)
```

```python
#6 - Tabuada, o usuario digita um numerp e o programa imprimi a tabuada do numero
# Usuário digita um número
numero = int(input("Digite um número: "))

# Imprime a tabuada do número
for i in range(1, 11):
    resultado = numero * i
    print(numero, "x", i, "=", resultado)
```

```python
#7 - Contar letras, o usuario diggita uma palavra e o programa conta as letras da palavra
# Usuário digita uma palavra
palavra = input("Digite uma palavra: ")

# Conta a quantidade de letras
quantidade = len(palavra)

# Mostra o resultado
print("A palavra tem", quantidade, "letras.")-
```

```python
#8 - Contar vogais, o usuario digita uma palavra e o programa conta as vogais
# Usuário digita uma palavra
palavra = input("Digite uma palavra: ")

# Variável para contar as vogais
contador = 0

# Percorre cada letra da palavra
for letra in palavra:
    if letra.lower() in "aeiou":
        contador += 1

# Mostra o resultado
print("A palavra tem", contador, "vogais.")
```

```python
#9 - Maior numero digitado, o usuario digita quatro numeros e o programa imprimi o maior numero digitado
# Usuário digita quatro números
n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))
n3 = int(input("Digite o terceiro número: "))
n4 = int(input("Digite o quarto número: "))

# Descobre o maior número
maior = max(n1, n2, n3, n4)

# Mostra o resultado
print("O maior número digitado foi:", maior)
```

```python
#10 - O usuario digita um numero entre 01 e 100 e o programa mostra quais são diviziveis por 3 e 5
 #Usuário digita um número entre 1 e 100
numero = int(input("Digite um número entre 1 e 100: "))

print("Números de 0 até", numero, "divisíveis por 3 e 5:")

# Percorre de 0 até o número digitado
for i in range(0, numero + 1):
    if i % 3 == 0 and i % 5 == 0:
        print(i)
```
